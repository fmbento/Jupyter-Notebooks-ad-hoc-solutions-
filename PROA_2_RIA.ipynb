# Search for Articles written by UA affiliated authors (@ua.pt) from UA's PROA Journals

### How-To
* Pack all the PDFs into a *PDFsPROA.zip* file, and upload it to the same folder / directory where this notebook is saved / running;
* Run this Notebook.

```python
# !pip install PyPDF2

import os
import shutil
import re

from PyPDF2 import PdfReader

!mkdir ./PDFs2RIA/

# Unpack the archive file
shutil.unpack_archive('./PDFsPROA.zip', './PDFs2RIA/', 'zip')

# folder path
dir_path = r'./PDFs2RIA'

res = []

# Iterate directory
for path in os.listdir(dir_path):
    # check if current path is a file
    if os.path.isfile(os.path.join(dir_path, path)):
        res.append(path)

# print(res)        

for pdf_file_path in res:
  
  print()
  print()
  print(pdf_file_path)
  print()
  obj = PyPDF2.PdfReader(dir_path + '/' + pdf_file_path)

  pgno = len(obj.pages)

  S = "@ua.pt"

  for i in range(0, pgno):
      PgOb = obj.pages[i] 
      Text = PgOb.extract_text()
      if re.search(S,Text):
          print("@ua.pt found on Page: " + str(i))

!rm ./PDFsPROA.zip
!rm -rf ./PDFs2RIA/
```
